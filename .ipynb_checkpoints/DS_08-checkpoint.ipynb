```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df.isnull().sum()
```

```python
df1 = df.drop(['deck'], axis=1 ,)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1['embarked'] = df1['embarked'].fillna(" ")
df1['embark_town'] = df1['embark_town'].fillna("")

```

```python
df1
```

```python
df1.isnull().sum()
```

##**Distribution Plots**

> Indented block



### **a) DistPlot**

```python
sns.distplot(x = df1['age'], bins = 10)
plt.show()
```

```python
sns.distplot(df1["age"], bins = 10,kde=False)
```

##**b) JOINTPLOT**

####**plot1 :- Scatter**
####**plot2 :- hex**

```python
sns.jointplot(x = df1["age"], y = df1["fare"], kind = "scatter")
sns.jointplot(x = df1["age"], y = df1["fare"], kind = "hex")
```

## **c) RUG PLOT**

```python
sns.rugplot(df1["fare"])
```

##**Categorical Plots**

###**a) BarPlot**

```python
sns.barplot(x="sex", y="age", data=df1, palette="Set1")
```

```python
sns.barplot(x="sex", y="age", data=df1, estimator=np.std,palette="Set2")
```

###**b) Count Plot**

```python
sns.countplot(x="sex", data=df1, palette='Set3')
```

### **c) Box Plot**

```python
sns.boxplot(x="sex", y="age", data=df1, palette="BuGn")
```

```python
sns.boxplot(x="sex", y="age", data=df1, hue="survived")
```

### **d) Voiline Plot**

```python
sns.violinplot(x="sex", y="age", data=df1, palette='Set1')
```

```python
sns.violinplot(x="sex", y="age", data=df1, hue="survived")
```

## **Advanced Plots**

### **a) Strip Plot**

```python
sns.stripplot(x="sex", y="age", data=df1, jitter=False, palette = 'Set2')
```

```python

sns.stripplot(x="sex", y="age", data=df1, jitter=True, palette = 'Set2')
```

```python
sns.stripplot(x="sex", y="age", data=df1, jitter=True, palette = 'Set2', hue='survived')
```

### **b) Swarm Plot**

```python
sns.swarmplot(x="sex", y="age", data=df1,palette = 'Set2')
```

```python
sns.swarmplot(x="sex", y="age", data=df1, palette = 'Set2', hue='survived')
plt.show()
```

## **Matrix Plots**

### **a) Heat Map**

```python
df1_encoded = pd.get_dummies(df1)
corr = df1_encoded.corr()

# Plot heatmap
sns.heatmap(corr)
```
