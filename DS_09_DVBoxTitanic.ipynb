```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset("titanic")
```

```python
sns.displot(df['age'],bins=10,kde=False)
```

```python
sns.jointplot(x=df['age'],y=df['fare'],kind='hex')
```

```python
sns.rugplot(df['fare'])
```

```python
sns.boxplot(x=df['sex'],y=df['age'],hue=df['survived'])
```

```python
sns.barplot(x=df['sex'],y=df['age'],estimator=np.var)
```

```python
sns.countplot(x=df['sex'],color='red')
```

```python
sns.violinplot(x=df['sex'],y=df['age'],hue=df['survived'])
```

```python
sns.stripplot(x=df['sex'],y=df['age'],jitter=False,hue=df['survived'])
```

```python
sns.swarmplot(x=df['sex'],y=df['age'],hue=df['survived'])
```
